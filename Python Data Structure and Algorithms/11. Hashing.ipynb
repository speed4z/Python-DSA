### Hashing :
A method of sorting and indexing data.

Hashing is to allow large data to be indexed using keys commonly created by formulas

#### Need of Hash:


#### Hash Function:    
    -A function that can be used to map arbitary size of data to data of a fixed size

#### key:        
    -Input data by a user
    

#### Hash Value:    
    -A value that is returned by Hash Function
    
#### Hash Table:    
    -It is a data structure which implements associative arrat abstract data type,
    a structure that can map keys to values

#### Collision:
    -A collision occurs when 2 different keys to hash function produces same output
    
#### Hash Functions :
    -mod function   
    -ASCII function 
    
#### Properties of Good Hash Functioon:
    -It distributes hash values uniformly across hash tables
    -It has to use all the input data
    

#### Collision resolution techniques:


Direct Chaining : implements the bucket as linked list.
    Colliding elements are stored in this list.
    
Oper Addressing : Colliding elements are stored in vacant buckets. During
    storage and lookup these are found through so called probing
    
    Linear Probing : It places new key into closest following empty cell.
    
    Quadratic Probing : Adding arbitary quadratic polynomial to the index
    until an empty cell is found.
        
    Double Hashing : Interval between probes is computed by another hash
    function
        

#### Hash Table is FULL:
    -For Direct Chaining: this situation never arrives
    -For Open Addressing: Create 2X size of current Hash Table and recall
    hashing for current keys
