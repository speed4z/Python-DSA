## Data Structures 
are different ways of organizing data that can be used effectively.

## Algorithms:
Set of rules and steps for a program to accomplish a task
should have:
    
1.Correctness

2.Efficiency

Flow to solve problem:
    input -> data structures and algorithms -> output

## Types of Data Structures:
Primitive and Non Primitive

Primitive :built in programming language itself
    
    int,float,char/string,boolean
    
    
Non Primitive: derrived from one or more primitive DS
    
    Linear: 
        
        Static => array
        Dynamic => Linked list, Stack, Queue
    Non Linear:
        
        Tree,Graph            

## Types of Algorithms:

-Simple recursive : Function calls itself (maybe with different parameters)

-Divide and Conquer : Divide the problem into smaller subproblem of same type;
    and solves subproblems recursively. Combine the solutions to the subproblem
    into a Solution to the original problem.
    eg. Merge sort,quick sort

-Dynamic Programming :  Work based on memorization; to find best solution
    generally used for optimization problems

-Greedy : We take the best we can without worrying future consequences;
    We hope choosing local optimum solution at each step will work at 
    global optimum solution.
    Find best solution, can be used for optimization problems;

-Brute Force : Try all possibilities until a satisfactory solution is found.

-Randomized : Use a random number at least once during the computation to make
    a decision. eg. Quick Sort.
