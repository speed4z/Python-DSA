### Sorting:
    -Arrangement of data in particular format.
    
#### Type of sorting algorihms:
    -Space used:
        -in Placed: Does not require any extra space for sorting.
        eg. bubble Sort
        
        -Out Placed: Requires extra space for sorting.
        eg. Merge Sort
        
    -Stability:
        -Stable: If algorithm does not change the sequence of similar
        content in which they appear, then it is Stable.
        eg. Insertion sort
        
        -Unstable: If algorithm changes the sequence of similar content
        in which they appear, then it is unstable
        eg. Quick Sort

#### Sorting Terminologies :
    - Increasing order
    - Decreasing order
    - Non-Increasing order
    - Non-Decreasing order
    
#### Algorithms:
    -Bubble sort
    -Selection sort
    -Insertion sort
    -Bucket sort
    -Merge sort
    -Quick sort
    -Heap sort
    
#### Which one to use:
    -Stability
    -Space Efficient
    -Time Efficient

### Bubble Sort:


```python
def bubble_sort(arr):    #-----O(N*N)
    for i in range(len(arr)-1):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    print(arr)

list1 = [2,3,2,1,5,23,153,234]
bubble_sort(list1)
```

### Selection Sort


```python
def selection_sort(arr):      #-----O(N*N)
    for i in range(len(arr)):
        min_index = i
        for j in range(i+1,len(arr)):
            if arr[min_index]>arr[j]:
                min_index = j
        arr[i],arr[min_index]=arr[min_index],arr[i]
    print(arr)

list1 = [2,3,2,1,5,23,153,234]
selection_sort(list1)
                
```

### Insertion Sort


```python
def insertion_sort(arr):      #-----O(N*N)
    for i in range(1,len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    print(arr)    
        
list1 = [2,3,2,1,5,23,153,234]
insertion_sort(list1)    
```

### Bucket sort


```python
#Number of buckets = sqrt(number of elements)
import math

def bucket_sort(arr):
    numofbuckets = round(math.sqrt(len(arr)))
    maxval = max(arr)
    buckets = []
    
    for i in range(numofbuckets):
        buckets.append([])
        
    for j in arr:
        index_b = math.ceil(j*numofbuckets/maxval)
        buckets[index_b-1].append(j)
    
    for i in range(numofbuckets):
        buckets[i] = insertion_sort(buckets[i])
    
    k = 0
    print(type(buckets[k]))
    for i in range(numofbuckets):
        for j in range(len(buckets[i])):
            arr[k] = buckets[i][j]
            k += 1
    return arr      
    
    
list1 = [2,3,2,1,5,23,20,18,10]
print(bucket_sort(list1))       
```

### Merge Sort:
    -Divide and conquer!


```python
def merge(arr,l,m,r):    #merge function : O(n)
    n1 = m-l+1
    n2 = r-m
    
    L = [0]*n1
    R = [0]*n2
    
    for i in range(n1):
        L[i] = arr[l+i]
    for j in range(n2):
        R[j] = arr[m+1+j]
        
    i = 0
    j = 0
    k = l
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i+=1            
        else:
            arr[k] = R[j]
            j+=1            
        k+=1
    
    while i < n1:
        arr[k] = L[i]
        i+=1
        k+=1
        
    while j < n2:
        arr[k] = R[j]
        j+=1
        k+=1

def merge_sort(arr,l,r):    # ----------- O(nlog n)
    if l < r:
        m=(l+(r-1))//2
        merge_sort(arr,l,m)
        merge_sort(arr,m+1,r)
        merge(arr,l,m,r)    
    return arr
    
list1 = [2,3,2,1,5,23,153,234]
merge_sort(list1,0,len(list1)-1)    
```
