### Doubly Linked List or DLL
Creation of Circular Singly Linked List CSLL:

    -Start
    -node value
    -create blank node , node.value = node value
    -node.next = NULL, node.prev = NULL
    -head = node , tail = node
    -Terminate

```python
#Create DLL doubly Linked  List :
class Node:
    def __init__(self,value = None):
        self.value = value
        self.next = None
        self.prev = None
        
class DLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def __iter__(self):   #To iterate the Linked List 
        node = self.head
        while node:
            yield node
            node = node.next 
```

```python
#Create a DLL
def create_DLL(self, nodeValue):         #O(1)
    node = Node(nodeValue)
    node.prev = None
    node.next = None
    self.head = node
    self.tail = node
    return "DLL created with node = ",nodeValue
    
DLL.create_DLL=create_DLL
```

#### insertion in DLL:


```python
def insert_DLL(self,value,loc):             #O(n)
    newNode = Node(value)

    if self.head is None:
        print("List not Found")

    else:
        if loc == 0:    #Beginning of LL  ===== O(1)  
            newNode.next = self.head
            newNode.prev = None
            self.head.prev = newNode
            self.head = newNode            

        elif loc == -1:  #End of LL  ====== O(1)             
            newNode.next = None
            newNode.prev = self.tail
            self.tail.next = newNode
            self.tail = newNode

        else:             #At middle of LL===== O(n)
            tempNode = self.head
            index = 0
            while index < loc-1:
                tempNode = tempNode.next
                index+=1
            nextNode = tempNode.next
            tempNode.next = newNode
            newNode.next = nextNode
            
            newNode.prev = tempNode
            nextNode.prev = newNode
            
DLL.insert_DLL = insert_DLL
```

```python
#Create a Doubly Linked list
DLL2 = DLL()
node1 = Node(101)
node2 = Node(102)
DLL2.head=node1
DLL2.head.next=node2
DLL2.tail=node2
DLL2.tail.prev=node1
```

```python
print_DLL(DLL2)

print("After insertion")
DLL2.insert_DLL(105,-1)

DLL2.insert_DLL(103,2)
DLL2.insert_DLL(104,3)
DLL2.insert_DLL(105,4)
DLL2.insert_DLL(106,5)
DLL2.insert_DLL(107,6)
DLL2.insert_DLL(108,7)

print_DLL(DLL2)
print("After insertion at start")
DLL2.insert_DLL(100,0)
print_DLL(DLL2)
```

```python
def traverse_DLL(self):          #O(n)
    if self.head is None:
        return "List doesn't exist"
    else:
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
DLL.traverse_DLL=traverse_DLL
```

```python
DLL2.traverse_DLL()
```

```python
def rev_traverse_DLL(self):          #O(n)
    if self.head is None:
        return "List doesn't exist"
    else:
        temp = self.tail
        while temp:
            print(temp.value)
            temp = temp.prev
    
DLL.rev_traverse_DLL=rev_traverse_DLL
```

```python
DLL2.rev_traverse_DLL()
```

```python
#Searching in the DLL          O(n)
def search_DLL(self,val):
    if self.head is None:
        return "List not Found"
    else:
        temp = self.head
        index = 0
        while temp:
            if temp.value == val:
                print("Value found at index: ",index)
                return index
            index += 1
            temp = temp.next
        return "Value not found in List"

DLL.search_DLL=search_DLL
```

```python
print(DLL2.search_DLL(200))
print(DLL2.search_DLL(103))
```

#### Deletion operation in DLL


```python
#Deletion operation in DLL       O(n)
def delete_DLL(self,loc):
    if self.head is None:
        return "List not Found"
    else:
        if loc == 0:
            if self.head == self.tail:
                self.head = None
                self.tail = None
            else:
                self.head = self.head.next
                self.head.prev = None
                
        elif loc == -1:
            if self.head == self.tail:
                self.head = None
                self.tail = None
            else:
                self.tail = self.tail.prev
                self.tail.next = None
        
        else:
            temp = self.head
            index = 0
            while index < loc-1:
                temp = temp.next
                index += 1
            temp.next = temp.next.next
            temp.next.prev = temp
        print("Deletion successful")

DLL.delete_DLL=delete_DLL
```

```python
print_DLL(DLL2)

print("After Deleting index 4")
DLL2.delete_DLL(4)
print_DLL(DLL2)

print("After deleting first and last element")
DLL2.delete_DLL(0)
DLL2.delete_DLL(-1)
print_DLL(DLL2)
```

```python
#deleting entire DLL:
def delete_entire_DLL(self):      #O(n)
    if self.head == None:
        print("DLL does not exists")
    else:
        temp = self.head
        while temp:
            temp.prev = None
            temp = temp.next
        self.head = None
        self.tail = None            
        
DLL.delete_entire_DLL = delete_entire_DLL   
```

```python
DLL2.delete_entire_DLL()
print_DLL(DLL2)
```
