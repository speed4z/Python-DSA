### Dictionary:
    -Dictionaries are used to store data values in key:value pairs.
    -A dictionary is a collection which is ordered*, changeable and does
    not allow duplicates.
    
*As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

```python
#Travesrse through Dictionary O(N)

#by keys
for key in b.keys():
    print(key,b[key])
print("\n")
    
#by values
for value in b:
    print(value)
print("\n")
    
#    OR
    
for value in b:
    print(value)
print("\n")

#Byt both key and value
for key,value in b.items():
    print(key,value)
```

```python
#Search in the Dictionary Average: O(1) worstcase: O(N)

def seach_dict(dict1,val):
    for key in dict1:
        if dict1[key] == val:
            return key,val
    return "Not Found"

dict1 = {12:11,32:22,34:33,23:44,34:23}

print(seach_dict(dict1,22))
print(seach_dict(dict1,100))
```

```python
#Deletion

dict3 = {1:2,2:3,3:4}
print(dict3)

#del dictname[keyname] => removes key-value pair with key as keyname
del dict3[1]
print(dict3)

#clear
#dictname.clear() Clears all elements from dictionary and makes it empty.
dict3.clear()
print("After clear method, dict3=",dict3)

#del dictname => deletes whole dictionary
del dict3
print(dict3) #error dict3 is not defined
```

### Dictionary Methods


```python
#copy()
dict4 = {"name":"abc","age":20}
newdict = dict4.copy()

print(dict4)
print(newdict)
```

```python
#fromkeys() #create dict with given keys and value set to None
newdict = {}.fromkeys([1,2,3])
print(newdict)
```

```python
#get()
print(dict4.get('name'))
#default value can be provided in get if key is not found in dict
print(dict4.get('address',"India"))
```

```python
#items()
#returns list of dict items; similar for keys and values
print(dict4.items())
```

```python
#keys
print(dict4.keys())
```

```python
#values()
print(dict4.values())
```

```python
#popitem()
#pops the pair from dict

#setdefault()
print(dict4.setdefault('name','john'))
print(dict4.setdefault('name1','john'))
```
