### BST:
    -In the left subtree, value is less than or equal to root node.
    -In the left subtree, value is greater than root node.

```python
#Create BST:                  #---------O(1)
class BSTNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

```python
#Insert Node in BST:           #-----------O(log N)
def insert_BST(rootNode,value):
    if rootNode.data == None:
        rootNode = value
    elif value <= rootNode.data:
        if rootNode.left is None:
            rootNode.left = BSTNode(value)
        else:
            insert_BST(rootNode.left,value)
            
    elif value > rootNode.data:
        if rootNode.right is None:
            rootNode.right = BSTNode(value)
        else:
            insert_BST(rootNode.right,value)  
    return "Successfully Inserted!"        
    
```

```python
BST1 = BSTNode(100)
insert_BST(BST1,5)
insert_BST(BST1,66)
insert_BST(BST1,20)
insert_BST(BST1,50)
insert_BST(BST1,26)
insert_BST(BST1,92)
```

```python
'''
Traversal of Binary Tree:
    -Depth first search :
        -Preorder
        -Inorder
        -Postorder
    -Breadth first search:
        -Level order traversal
'''
#PREORDER Traversal
def preorderTraversal(rootNode):      # O(n)
    if not rootNode:
        return
    print(rootNode.data)
    preorderTraversal(rootNode.left) #======O(n/2)
    preorderTraversal(rootNode.right)#======O(n/2)
```

```python
#INORDER Traversal
def inorderTraversal(rootNode):      # O(n)
    if not rootNode:
        return    
    inorderTraversal(rootNode.left) #======O(n/2)
    print(rootNode.data)
    inorderTraversal(rootNode.right)#======O(n/2)
```

```python
#POSTORDER Traversal
def postorderTraversal(rootNode):      # O(n)
    if not rootNode:
        return    
    postorderTraversal(rootNode.left) #======O(n/2)    
    postorderTraversal(rootNode.right)#======O(n/2)
    print(rootNode.data)
```

```python
#LEVELORDER Traversal
def levelorderTraversal(rootNode):#-------O(n)
    if not rootNode:
        return "Tree not found"
    qu = []
    qu.append(rootNode)
    while len(qu)>0: #-------------O(n)
            
        root = qu.pop(0)
        print(root.data)    # .data as in TreeNode.data
        
        if (root.left is not None):
            qu.append(root.left)
            
        if (root.right is not None):
            qu.append(root.right)
```

```python
#SEARCH in Binary Search Tree       #---------O(log n)
def search_BST(rootNode,value):
    if rootNode.data == value:
        print("Value is Found")
        return rootNode.data
    elif value < rootNode.data and rootNode.left is not None:
        if rootNode.left.data == value:
            print("Value is Found")
            return rootNode.left.data
        else:
            search_BST(rootNode.left,value)
            
    elif value > rootNode.data and rootNode.right is not None:
        if rootNode.right.data == value:
            print("Value is Found")
            return rootNode.right.data
        else:
            search_BST(rootNode.right,value)
```

#### Delete From BST:


```python
def minValueNode(Node):
    current = Node
    while current.left is not None:
        current = current.left
    return current
```

```python
def deleteNode(rootNode,nodeValue):  #-------space and time :O(log N)
    if rootNode is None:
        return rootNode
    if nodeValue < rootNode.data:
        rootNode.left = deleteNode(rootNode.left,nodeValue)
    elif nodeValue > rootNode.data:
        rootNode.right = deleteNode(rootNode.right,nodeValue)
    else:
        if rootNode.left is None:
            temp = rootNode.right
            rootNode = None
            return temp
        
        if rootNode.right is None:
            temp = rootNode.left
            rootNode = None
            return temp
    
        temp = minValue(rootNode.rightChild)
        rootNode.data = temp.data
        rootNode.right = deleteNode(rootNode.right,temp.data)
    return rootNode
```

```python
def deleteBST(rootNode):
    rootNode.data = None
    rootNode.right = None
    rootNode.left = None
    return "The entire BST has been deleted. "
```
