### Big O
is the language and metric we use to describe the efficiency of the algorithms

Types:
    O(N), O(N^2), O(N^3)

#### best case
    -Big Omega => complexity at least more than BEST CASE.
#### Average case
    -Big Theta => complexity within bounds of WORST and BEST CASE.
#### worst case
    -Big O => complexity less of equal than WORST CASE.
