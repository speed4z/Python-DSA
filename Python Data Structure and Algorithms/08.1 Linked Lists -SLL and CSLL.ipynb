### Linked list:
    -a form of sequential collection and it does not have to be in order.
    A linked list is made up of independent nodes that may contain any type
    of data and each node has areference to the next node in the link.

### Linked list vs Arrays:
    -Elements of Linked List are independent object.
    -Linked list is dynamic DS, size is not defined at start.
    -Memory allocation iis not continuous in LL.
    -Insertion and removal is easy in LL.
    -Random element access is efficient in array (coz of indexing).

### Types of Linked List
    -Singly LL
    -Circular LL
    -Doubly LL
    -Circular Doubly LL

### Singly Linked List

```python
#Create SLL Singly Linked List:
class Node:
    def __init__(self,value = None):
        self.value=value
        self.next=None
        
class SLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def __iter__(self):   #To iterate the Linked List 
        node = self.head
        while node:
            yield node
            node = node.next 
```

#### Insert in Singly Linked List


```python
def insert_SLL(self,value,loc):
    newNode = Node(value)

    if self.head is None:
        self.head = newNode
        self.tail = newNode

    else:
        if loc == 0:    #Beginning of LL  ===== O(1)  
            newNode.next = self.head
            self.head = newNode

        elif loc == -1:  #End of LL  ====== O(1)             
            newNode.next = None
            self.tail.next = newNode
            self.tail = newNode

        else:             #At middle of LL===== O(n)
            tempNode = self.head
            index = 0
            while index < loc-1:
                tempNode = tempNode.next
                index+=1
            nextNode = tempNode.next
            tempNode.next = newNode
            newNode.next = nextNode
            
SLL.insert_SLL = insert_SLL
```

```python
def search_SLL(self,nodeValue):
    if self.head is None:
        return "This is empty list"
    else:
        node = self.head
        loc = 0
        while node is not None:
            if node.value == nodeValue:
                print("index of value in LL is ",loc)
                return loc
            loc += 1
            node = node.next
        return "Value not found"                

SLL.search_SLL=search_SLL
```

#### Deletion operation in Linked list


```python
def deleteNode(self,location):
    if self.head is None:
        print("Empty List")
    else:
        if location == 0:        #First Node deletion == O(1)           
            if self.head == self.tail:
                self.head = None
                self.tail = None
            else:
                self.head = self.head.next
                
        elif location == -1:      #Last Node deletion == O(1)
            if self.head == self.tail:
                self.head = None
                self.tail = None
            else:                  #Middle Node deletion == O(n)
                node = self.head
                while node is not None:
                    if node.next == self.tail:
                        break
                    node = node.next
                node.next = None
                self.tail = node
                
        else:
            temp = self.head
            index = 0
            while index < location - 1:
                temp = temp.next
                index += 1
            nextnode = temp.next
            temp.next = nextnode.next
            nextnode.next = None  
                   
    
SLL.deleteNode = deleteNode
```

### Circular Linked List

Creation of Circular Singly Linked List CSLL:

    -Start
    -node value
    -create blank node , node.value = node value
    -node.next = node
    -head = node , tail = node
    -Terminate

```python
#Create SLL Single Linked  List and Print it:
class Node:
    def __init__(self,value = None):
        self.value=value
        self.next=None
        
class CSLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def __iter__(self):   #To iterate the Linked List 
        node = self.head
        while node:
            yield node
            node = node.next
            if node.next == self.head: #Condition to exit infinite loop
                break
            node = node.next
```

```python
def create_CSLL(self,val):
    node = Node(val)
    node.next = node
    self.head = node
    self.tail=node
    return "CSLL has been created for value: "+str(val)

CSLL.create_CSLL = create_CSLL
```

```python
CSLL1 = CSLL()
CSLL1.create_CSLL(21)
```

```python
def print_CSLL(LL):
    temp = LL.head
    if temp is None:
        return "Empty list"
    else:
        while True:
            print(temp.value)
            if temp.next == LL.head:
                break
            temp = temp.next
```

#### inserting in CSLL


```python
def insert_CSLL(self,val,loc): # ===== O(n)
    if self.head is Node:
        return "CSLL does not exist"
    else:
        newNode = Node(val)
        if loc == 0:
            newNode.next = self.head
            self.head = newNode
            self.tail.next = newNode
        
        elif loc == -1:
            newNode.next = self.head
            self.tail.next = newNode
            self.tail = newNode
        
        else:
            temp = self.head
            index = 0
            while index < loc -1:
                temp = temp.next
                index += 1
            nextNode = temp.next
            temp.next = newNode
            newNode.next = nextNode
            
        return "Inserted Successfully"

CSLL.insert_CSLL=insert_CSLL
```

```python
# Search in CSLL

def search_CSLL(self,val):
    if self.head == None:
        return "Empty list"
    else:
        temp = self.head
        index = 0
        while True:
            if(temp.value == val):
                print("Index of val in CCSLL is :",index)
                return index
            temp = temp.next
            index += 1
            if temp == self.tail:
                return "Value not Found"
                break

CSLL.search_CSLL=search_CSLL
```

#### Deleting from CSLL:


```python
def delete_CSLL(self,loc): # ===== O(n)
    if self.head is Node:
        return "CSLL does not exist"
    else:
        if loc == 0:
            if self.head == self.tail:
                self.head = None
                self.tail = None
            else:                
                self.head = self.head.next
                self.tail.next = self.head
        
        elif loc == -1:
            if self.head == self.tail:
                self.head = None
                self.tail = None
            else:
                temp = self.head
                while temp is not None:
                    if temp.next == self.tail:
                        break
                    temp=temp.next
                temp.next = self.head
                self.tail = temp
                
        else:
            temp =self.head
            index = 0
            while index < loc-1:
                temp = temp.next
                index += 1
            nextNode = temp.next
            temp.next = nextNode.next
                

CSLL.delete_CSLL=delete_CSLL
```

```python
#Delete entire CSLL
def delete_entire_CSLL(self):
    if self.head == None:
        print("SLL does not exists")
    else:
        self.head = None
        self.tail = None

CSLL.delete_entire_CSLL=delete_entire_CSLL   
```
