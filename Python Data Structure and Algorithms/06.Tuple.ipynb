### Tuple:
    -A tuple is an immutable sequence of python object.
    -Tuples are comparable and hashable.
    -Tuple values are ordered and duplicates are allowed.

```python
#create Tuple
t1 = ('a','b','c')
print(t1)
print(type(t1))

#Not a tuple example
t2=('a')
print(t2)
print(type(t2))

#To Correct the above exampe
#Not a string but a tuple
t3 = ('a',)
print(t3)
print(type(t3))

#OR
t4 = tuple('a')
print(t4)
print(type(t4))

t5 = 'a','b','v'
print(t5)
print(type(t5))

```

```python
tup = ('a','b','c','d','e')
#Tuple values are stored in continuous memory

#Access Values
print(tup[0])
print(tup[-1])
#slice
print(tup[:-1])
print(tup[1:3])
```

```python
#Tuple is Immutable 
#hence
tup[2]='z'
#is not allowed
```

```python
#Traversing a Tuple : complexity O(N)
for i in tup:
    print(i,end=" ")
print('')
#OR
for i in range(len(tup)):
    print(tup[i],end= " ")
```

```python
#Search in Tuple : Complexity O(N)
#For Sorted tule, binary search costs O(log N)
def search_tuple(tup,val):
    for i in range(len(tup)):
        if tup[i]==val:
            return tup.index(val)
    return "Not Found"

print(search_tuple(tup,'c'))
print(search_tuple(tup,'n'))
```

### Tuple Vs List:
    -Lists are mutable but Tuples are immutable.
    
Functions that can be used for both list and tuples:

    -len()
    -max()/min()
    -sum()
    -any() / all()
    -sorted()
    
Methods that can be used for both list and tuples:

    -count()
    -index()
    
Methods that can be used for list but NOT for tuples:

    -append(), insert(),remove(),pop(),clear(),sort(),reverse()
