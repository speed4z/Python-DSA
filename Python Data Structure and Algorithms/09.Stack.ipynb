### Stack:
Stack is data structure which stores items in a Last In First Out manner

#### Stack Operations:
    -create stack
    -push
    -pop
    -peek
    -isEmpty # returns True or False
    -isFull  # returns True or False
    -delete stack

Stack using List:

    -easy to implement
    -Speed problem when it grows
    
Stack using Linked List:

    -Fast Performance
    -Implementaion is not easy

Stack using deque:
    
    -easy to implement

### When to Use:
    -LIFO Implementation
    -Minimum data Curruption
    -Random Access is not required

```python
from collections import deque
```

```python
#Stack using deque:
class stack:
    def __init__(self):
        self.container = deque()
        
    def push(self,val):
        self.container.append(val)
        
    def pop(self):
        return self.container.pop()
    
    def peek(self):
        return self.container[-1]
    
    def isEmpty(self):
        return len(self.container)==0
    
    def size(self):
        return len(self.container)
```

```python
#Print_stack method
def print_stack(stk):
    for i in range(len(stk.container)-1,-1,-1):
        print(stk.container[i])
```

```python
#create stack
stk = stack()
stk.push(29)
stk.push(86)
stk.push(45)
stk.push(63)
stk.push(42)
stk.push(23)
stk.push(43)

print_stack(stk)
```

```python
print(stk.pop())
print("After pop")
print_stack(stk)
```

```python
print(stk.peek())
print("After peek")
print_stack(stk)
```

```python
print(stk.isEmpty())
```

```python
print(stk.size())
```
